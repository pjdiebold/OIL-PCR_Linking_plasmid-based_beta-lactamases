```python
df=pd.read_csv("combined_spikein_curated.txt", sep='\t')
df.head()
```

```python
df2=df
df2["total"]=df2['sum_size'].groupby(df2['num']).transform('sum')
df2['per']=100*df2['sum_size']/df2['total']
df2.to_csv("mid_out",sep='\t')
```

```python
df2.head()
```

```python
cut_thresh=0.5
cut_list=[]
for row in df2.iterrows():
    per=row[1]['per']
    if per > cut_thresh:
        cut_list.append(1)
    else:
        cut_list.append(0)
df2['cut_off']=cut_list

df2.head()
```

```python
rep_list=['Donor','percent ','fiull_tax']
count_df=df2.groupby(rep_list)['cut_off'].aggregate('sum').rename("detections").reset_index()
detect_df=pd.merge(df2, count_df, on=rep_list)
#multi_df=detect_df[(detect_df.exp == 'multi') & (detect_df.detections == 4)]
#single_df=detect_df[(detect_df.exp == 'Single') & (detect_df.detections == 3)]
#exception_df=detect_df[(detect_df.exp == 'Single') & (detect_df.time==2) & (detect_df.gene=='CTX') & (detect_df.rep != 'A') & (detect_df.detections >= 2)]
#tran_df=detect_df[(detect_df.exp == 'transposon') & (detect_df.detections == 3)]
filter_df=detect_df[detect_df['detections']==3]
filter_df=filter_df[detect_df['blast_hit'] == 'CmR']
filter_df=filter_df[detect_df['percent '] != 'pSpike']
filter_df.to_csv('filtered_detections.txt',sep='\t')


#trimmed_df=pd.concat([multi_df, single_df,exception_df,tran_df])
#trimmed_df.head()
```
